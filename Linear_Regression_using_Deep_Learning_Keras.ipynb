### Build linear model on toy data - boston data

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.datasets import boston_housing
```

### Check the shape of datasets

```
(x_train,y_train),(x_test,y_test) = boston_housing.load_data()
```

```
type(x_train)
```

```
print('shape of train data',x_train.shape)
print('shape of test data',x_test.shape)
print('no.of features in train data',x_train[0,:])
print('sample data in train set',y_train[0])
```

```
nFeatures = x_train.shape[1]

model = Sequential()
model.add(Dense(1,input_shape=(nFeatures,),activation='linear'))

model.compile(optimizer='rmsprop',loss='mse',metrics=['mse','mae'])
```

```
model.summary()
```

```
model.fit(x_train,y_train,batch_size=2,epochs=500,verbose=True)
```

```
model.evaluate(x_test,y_test,verbose=True)
```

```
pred = model.predict(x_test)

print('true values',y_test[:5])
print('predicted values',pred[:5,0])
```
